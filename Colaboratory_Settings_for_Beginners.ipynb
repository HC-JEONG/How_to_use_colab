# Colaboratory?

> Jupyter를 통해 Python 코딩한 것을 **구글 가상 머신**에서 실행시켜준다.

따라서, **Jupyter를 설치하지 않아도** 인터넷만 되면 Jupyter 노트 환경에서 Python 코드를 코딩하고 결과를 출력할 수 있다.

[Colaboratory FAQ](https://research.google.com/colaboratory/faq.html)



## Colaboratory의 몇 가지 기능

> 기본적인 것은 기존 Jupyter와 비슷하다.

> [파일] 탭에는 아래와 같은 기능이 있다.

*   **새 python 3 노트**로 새 Jupyter 노트북을 만들 수 있다.
*   **노트 열기**의 **업로드**에서 내 컴퓨터(로컬 저장소)에 있는 파이썬 코드를 불러올 수 있고, **GOOGLE 드라이브**와 **GITHUB**에서도 각각 불러올 수 있다.
*   **드라이브**(구글 드라이브), **GITHUB**에 바로 저장 가능하다.
*   본인 컴퓨터에 저장하고 싶으면 **.ipynb**나 **.py**로 다운로드 하면 된다.

> [런타임] 탭에는 아래와 같은 기능이 있다.

*   모든 셀 또는 특정 셀을 실행할 수 있다.
*   런타임 중단, 재실행을 할 수 있다.
*   **런타임 유형 변경**에서 **하드웨어 가속기**를 선택 할 수 있는데, 구글 가상 머신의 **CPU**, **GPU**, **TPU** 중에서 선택할 수 있다.

> [도구] - [환경설정]에는 아래와 같은 기능이 있다.

*   **사이트**에서 테마를 지정할 수 있다.
*   **편집기**에서 노트에 대한 기본적인 설정(들여쓰기, 행 번호 표시 등)을 설정할 수 있다.
*   **기타**에서 파워 레벨을 바꾸면 수정할 때 [효과](https://twitter.com/GoogleColab/status/1112708634905964545)가 나타난다.

## matplotlib에 한글과 마이너스 기호 사용하기

> Colaboratory가 쉽게 말해 구글 컴퓨터의 일부를 할당 받아 내가 짠 코드를 실행하고 결과만 받는 것이기 때문에, matplotlib를 이용해 그래프를 만들면 한글폰트가 없어서 글씨가 깨진 채 보인다.

> 아래 코드는 한글 폰트를 사용하기 위한 과정이다.

> 나눔 폰트 설치

```
!apt-get update -qq
!apt-get install fonts-nanum* -qq
```

> Nanum으로 시작하는 폰트 이름과 설치 경로를 확인한다.

```
import matplotlib.font_manager as fm  # 폰트 관련 용도
sys_font=fm.findSystemFonts()
nanum_font = [f for f in sys_font if 'Nanum' in f]
nanum_font
```

> '/usr/share/fonts/truetype/nanum/'에 설치되어있는 것을 알 수 있다.

> 원하는 글씨체의 경로를 변수설정 해주고, fm.FontProperties를 이용해 설정해준다.

```
path = '/usr/share/fonts/truetype/nanum/NanumGothicEco.ttf'
# NanumGothicEco 자리에 다른 글씨체를 입력하면 다른 글씨체로 설치된다.

font_name = fm.FontProperties(fname=path, size=10).get_name()
print(font_name)

import matplotlib.pyplot as plt
plt.rc('font', family=font_name)
```

> 마지막으로 fm._rebuild()를 rebuild 해준다.

```
fm._rebuild()
```

> 마이너스 기호는 Jupyter에서 해결하는 방법과 같다.

```
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False
```

> 아래 테스트 코드를 실행하면 마이너스 기호와 한글 폰트가 보인다.

```
import numpy as np
plt.plot(np.random.randint(-5, 5, 10), np.random.randint(-5, 5, 10), 'bo--')
plt.title('테스트')
plt.xlabel('엑스')
plt.ylabel('와이')
plt.show()
```
